```python
import ipywidgets as widgets
from IPython.display import display
from PyPDF2 import PdfReader

def extract_text_from_pdf(button):
    # Get the input password and check it
    input_password = password_widget.value
    correct_password = '!#%246'

    if input_password == correct_password:
        output_widget.clear_output()
        with output_widget:
            try:
                print('Password Matched! 🔐')
                print('Access Granted! ✅')

                # Get the PDF file name and validate it
                pdf_name = pdf_name_widget.value
                reader = PdfReader(pdf_name)

                num_of_pages = len(reader.pages)
                start_page = start_page_widget.value - 1
                stop_page = stop_page_widget.value

                if 0 <= start_page < stop_page <= num_of_pages:
                    text = ''

                    for i in range(start_page, stop_page):
                        page = reader.pages[i]
                        text += '\n' + '_' * 59 + f'Page {i + 1}' + '_' * 59 + '\n\n\n' + page.extract_text() + '\n\n'

                    print('Text Extracted Successfully! 🎉\n')
                    print(text)
                else:
                    print(f'Error: Please enter valid page numbers (1-{num_of_pages})!')

            except Exception as e:
                print(f'Error: {str(e)}')
    else:
        output_widget.clear_output()
        with output_widget:
            print('Incorrect Password! 🚫 Access Denied! ❌')

# Create widgets
password_widget = widgets.Password(description='Enter Password:')
pdf_name_widget = widgets.Text(description='PDF Name:')
start_page_widget = widgets.IntText(value=1, description='Start Page:')
stop_page_widget = widgets.IntText(value=1, description='End Page:')
output_widget = widgets.Output()

# Create button
button = widgets.Button(description="Extract Text")
button.on_click(extract_text_from_pdf)

# Display widgets
display(password_widget, pdf_name_widget, start_page_widget, stop_page_widget, button, output_widget)

```
